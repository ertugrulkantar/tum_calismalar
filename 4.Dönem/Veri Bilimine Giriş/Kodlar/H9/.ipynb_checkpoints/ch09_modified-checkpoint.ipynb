### Figures and Subplots

```python
fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(2, 2, 1)
plt.plot(np.random.randn(30))
ax2 = fig.add_subplot(2, 2, 2)
plt.plot(np.random.randn(50), 'k')
ax3 = fig.add_subplot(2, 2, 3)
plt.plot(np.random.randn(50).cumsum(), 'k--')
ax4 = fig.add_subplot(2, 2, 4)
plt.plot(np.random.randn(100).cumsum(), 'k.')
plt.plot(np.random.randn(100).cumsum(), 'r--')
```

### Colors, Markers, and Line Styles

* Arguman olarak 2 array verirsen birisi x digeri y eksenini temsil eder.

```python
plt.plot(np.random.randn(30), np.random.randn(30), 'g--')
```

```python
plt.plot(np.random.randn(30), np.random.randn(30), linestyle='--', color='g')
```

* Cizgi grafikte noktalari gostermek istersen...

```python
plt.plot(np.random.randn(30).cumsum(), 'ko--')
```

```python
plt.plot(np.random.randn(30).cumsum(), color='k', linestyle='dashed', marker='o')
```

* Drawpost farkli bir gosterim sunar merdiven gibi...

* Etiketlemek icin legend kullanilir, sag ustteki bilgiler icin."best" parametresi yeri kendi ayarlar...

```python
data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
```

## Plotting with pandas

### Line Plots

```python
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
s.plot()
```

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0),
                  columns=['A', 'B', 'C', 'D'],
                  index=np.arange(0, 100, 10))
df.plot()
```

### Bar Plots

```python
fig, axes = plt.subplots(2, 1)
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))
data.plot.bar(ax=axes[0], color='k', alpha=0.7)
data.plot.barh(ax=axes[1], color='k', alpha=0.7)
```

```python
df = pd.DataFrame(np.random.rand(6, 4),
                  index=['one', 'two', 'three', 'four', 'five', 'six'],
                  columns=pd.Index(['A', 'B', 'C', 'D'], name='Genus'))
df
```

```python
df.plot.bar()
```

```python
df.plot.barh(stacked=True, alpha=0.5)
```
