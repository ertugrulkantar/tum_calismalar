# Data Loading, Storage, and File Formats

```python
import numpy as np
import pandas as pd
```

## Reading Data in Text Format

```python
df = pd.read_csv('examples/ex1.csv')
df
```

```python
pd.read_table('examples/ex1.csv', sep=',')
```

* Basligin olmadigi durumlarda...

```python
pd.read_csv('examples/ex2.csv', header=None)
```

* Sutun adlarini kendin vermek icin...

```python
pd.read_csv('examples/ex2.csv', names=['a', 'b', 'c', 'd', 'message'])
```

* Etiketi index olan sutun index(Row index) olarak ataniyor...

```python
names = ['a', 'b', 'c', 'd', 'message']
pd.read_csv('examples/ex2.csv', names=names, index_col='message')
```

* Iki tane row index atamak icin... Baya ise yarar bir seymis.

```python
parsed = pd.read_csv('examples/csv_mindex.csv',
                     index_col=['key1', 'key2'])
parsed
```

* Bazi satirlari atlamak icin...

```python
pd.read_csv('examples/ex4.csv', skiprows=[0, 2, 3])
```

```python
result = pd.read_csv('examples/ex5.csv')
result
```

```python
pd.isnull(result)
```

* Altta hangi degerlerin NULL olarak sayilacagini secebilirsin...

```python
result = pd.read_csv('examples/ex5.csv', na_values=['NULL'])
result
```

```python
result = pd.read_csv('examples/ex5.csv', na_values=[5,10,'foo'])
result
```
